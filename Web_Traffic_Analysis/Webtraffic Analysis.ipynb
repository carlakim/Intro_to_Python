__Background Story:__

Carla, Juyin and Xin are interns at L'Oréal. 

They want to compare visitors to the L'Oréal website.

Each of them collects site traffic data from different sources:

    Carla collects data for Google Analytics.
    Juyin collects data for Facebook Ads.
    Xin collects data for email campaigns.

---
#### Your task:

You are give three text files:

    "Cagla.txt"
    "Juyin.txt"
    "Xin.txt"

Each file contains 15 values (number of visits) collected every two days.

Steps to follow:

* Read the data from each file and store the values __in three separate lists__. (One for Carla, one for Juyin and one for Xin).
* <span style="color:crimson">Note:</span> if you are reading data from a text file as a list. The data in your list is of type text (string), you need to convert it to float. For that, you can loop over the values in the list and convert them to float, you can update the values in the original list or create a new one.
* Using matplotlib, plot the data as a line graph showing the two-day visits for each source __on the same plot__.
* Label the axes and include a legend.
 
Graph Specifications:

* X-axis: Day
* Y-axis: Number of visits
* Title: "L'Oréal daily visits"
* Each line should have a different color and line style.

```python
import matplotlib.pyplot as plt 
```

```python
#Read the data from each file and store the values in three separate lists. (One for Carla, one for Juyin and one for Xin).
c_file = open("./sitedata/Carla.txt", 'r')
carla_l = c_file.readlines() 
c_file.close()

j_file = open("./sitedata/Juyin.txt", 'r')
juyin_l = j_file.readlines() 
j_file.close()

x_file = open("./sitedata/Xin.txt", 'r')
xin_l = x_file.readlines() 
x_file.close()
```

```python
#make list values int 
def int_num (lista):
    l = []
    for i in range(len(lista)):
        l.append(int(lista[i]))
    return l
```

```python
c = int_num(carla_l)
j = int_num(juyin_l)
xi = int_num(xin_l)

print(c)
print(j)
print(xi)
```

```python
#list for x, the days 
x = list(range(1,16,1))
x
```

```python
#plot

fig,ax = plt.subplots(nrows=1, ncols=1, figsize=(8,3))

#lines
ax.plot(x,c, color = 'red', marker = '*',label='Google Analytics');
ax.plot(x,j, color = 'green', marker = 'o',linestyle = '--', label='Facebook Ads');
ax.plot(x,xi, color = 'royalblue', marker = 'x',linestyle = ':',label = 'Email Campaigns');

#details
ax.set_title("L'Oréal daily visits");
ax.set_xlabel("Day");
ax.set_ylabel("Number of visits");
plt.legend(bbox_to_anchor=(1.3, 0.8),loc='right')

ax.grid(alpha = 0.5)
```

```python
#Calculate the total visits by summing the visits from all three sources and add this as a fourth line in the plot.
tot = []
for i in range(len(c)):
    adi = c[i]+ j[i]+xi[i]
    tot.append(adi)

print(tot)
```

```python
#plot with extra line
fig,ax = plt.subplots(nrows=1, ncols=1, figsize=(9,5))

#lines
ax.plot(x,c, color = 'red', marker = '*',label='Google Analytics');
ax.plot(x,j, color = 'green', marker = 'o',linestyle = '--', label='Facebook Ads');
ax.plot(x,xi, color = 'royalblue', marker = 'x',linestyle = ':',label = 'Email Campaigns');
ax.plot(x,tot, color = 'purple', marker = 's',label = 'Total Visits');

#details
ax.set_title("L'Oréal daily visits");
ax.set_xlabel("Day");
ax.set_ylabel("Number of visits");
plt.legend(bbox_to_anchor=(1.3, 0.8),loc='right')

ax.grid(alpha = 0.5)
```
